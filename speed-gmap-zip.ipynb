```python
import gmaps

# Google developer API key
from config import gkey

# Access maps with unique API key
gmaps.configure(api_key=gkey)
```

```python
zipcodes = []
latList = []
longList = []
speedCamLoc = os.path.join("..", "Group-4-", "Speed_Zip_LL.csv")

with open(speedCamLoc, 'r', encoding='utf-8') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    header = next(csvreader)
    for row in csvreader:
        zipcode = row[1]
        latitude = row[2]
        longitude = row[3]
        zipcodes.append(zipcode)
        latList.append(latitude)
        longList.append(longitude)

#converting str into floats
f_lat = []
for item in latList:
    f_lat.append(float(item))

f_long = []
for item in longList:
    f_long.append(float(item))
```

```python
coordList = []       
coordList = list(zip(f_lat, f_long))
coordList
```

```python
figure_layout = {
    'width': '1800px',
    'height': '1600px',
    'border': '2px solid black',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig = gmaps.figure(layout=figure_layout)
```

```python
# Assign the marker layer to a variable
my_points = gmaps.marker_layer(coordList)
# Add the layer to the map
fig.add_layer(my_points)
fig
```
