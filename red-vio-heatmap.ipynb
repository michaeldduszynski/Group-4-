```python
heatmap_df = pd.read_csv(os.path.join("..", "Group-4-", "red-violation-longlat-indexmatch.csv"))
heatmap_df.dropna(inplace = True) 
heatmap_df.head()
```

```python
locations = heatmap_df[["LATITUDE", "LONGITUDE"]]
violations = heatmap_df["Total Violations"].astype(float)
```

```python
# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=violations, 
                                 dissipating=False, max_intensity=100,
                                 point_radius=.001, opacity=0.5, gradient=[(0, 255, 255, 0),(0, 255, 255, 1),(0, 191, 255, 1),(0, 127, 255, 1),
                                                              (0, 63, 255, 1),(0, 0, 255, 1),(0, 0, 223, 1),(0, 0, 191, 1),(0, 0, 159, 1),
                                                              (0, 0, 127, 1),(63, 0, 91, 1),(127, 0, 63, 1),(191, 0, 31, 1),(255, 0, 0, 1)])

# Add layer
fig.add_layer(heat_layer)
# Display figure
fig
```
